```python
import numpy as np
from time import time
```

```python
def strassen(A,B,n):
    if n ==2:
        
        M1= (A[0,0]+A[1,1])*(B[0,0]+B[1,1])
        M2 = (A[1,0]+A[1,1])*B[0,0]
        M3 = A[0,0]*(B[0,1]-B[1,1])
        M4 = A[1,1]*(B[1,0]-B[0,0])
        M5 = (A[0,0]+A[0,1])*B[1,1]
        M6 = (A[1,0]-A[0,0])*(B[0,0]+B[0,1])
        M7 = (A[0,1]-A[1,1])*(B[1,0]+B[1,1])

        C1 = M1+M4-M5+M7
        C2 = M3+M5
        C3 = M2+M4
        C4 = M1-M2+M3+M6

        

        return np.array([[C1,C2],[C3,C4]])

    
    else:

        half = int(n//2)
        A1 = A[:half,:half]
        A2 = A[:half,half:]
        A3 = A[half:,:half]
        A4 = A[half:,half:]
       

        B1 = B[:half,:half]
        B2 = B[:half,half:]
        B3 = B[half:,:half]
        B4 = B[half:,half:]





        Q1 = np.add(strassen(A1,B1,n/2),strassen(A2,B3,n/2))
        Q2 = np.add(strassen(A1,B2,n/2),strassen(A2,B4,n/2))
        Q3 = np.add(strassen(A3,B1,n/2),strassen(A4,B3,n/2))
        Q4 = np.add(strassen(A3,B2,n/2),strassen(A4,B4,n/2))

        return np.concatenate((np.concatenate((Q1,Q2),axis=1),np.concatenate((Q3,Q4),axis=1)),axis=0)
```
